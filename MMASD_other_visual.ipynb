```python
import os
import pickle
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from tqdm import tqdm

# ========== 基础配置（与训练代码完全一致） ==========
PKL_2D_DIR = '/root/second/2D'
PKL_3D_DIR = '/root/second/3D'
MODALITY = '3d'
MAX_FRAMES = 32
BATCH_SIZE = 4
ANOMALY_THRESHOLD = 0.8
NUM_JOINTS = 71
OUTPUT_DIR = 'asd_anomaly_detection_results_frame_level'  # 模型权重所在目录
TEST_SPLIT_RATIO = 0.2  # 测试集占比（和训练集同源划分）

# 中文显示配置
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 设备配置
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"使用设备：{device}")

# ========== 工具函数（复用训练代码中的核心函数） ==========
def get_pkl_files(dir_path):
    """获取目录下所有pkl文件"""
    return [os.path.join(dir_path, f) for f in os.listdir(dir_path) if f.endswith('.pkl')]

def get_all_themes(pkl_files):
    """从PKL文件中提取所有Theme名称"""
    themes = set()
    for file_path in pkl_files:
        with open(file_path, 'rb') as f:
            data = pickle.load(f)
        themes.add(data['metadata']['theme_name'])
    return sorted(list(themes))

# ========== 测试集专用数据集类（仅加载测试集数据） ==========
class AnomalySkeletonTestDataset(Dataset):
    """测试集数据集（仅加载并处理测试集部分）"""
    def __init__(self, pkl_files, modality='3d', target_theme=None, max_frames=32, test_ratio=0.2):
        self.modality = modality
        self.skeleton_key = 'skeleton_2d' if modality == '2d' else 'skeleton_3d'
        self.target_theme = target_theme
        
        # 加载该Theme的所有数据
        self.all_frame_data = []
        self.all_frame_info = []
        sample_idx = 0
        
        for file_path in pkl_files:
            with open(file_path, 'rb') as f:
                data = pickle.load(f)
            theme_name = data['metadata']['theme_name']
            if theme_name != target_theme:
                continue
            
            for sample in data['samples']:
                skeleton = sample[self.skeleton_key]
                motion_name = sample['motion_name']
                T = skeleton.shape[0]
                for frame_idx in range(T):
                    frame = skeleton[frame_idx]
                    self.all_frame_info.append({
                        'sample_idx': sample_idx,
                        'motion_name': motion_name,
                        'frame_idx': frame_idx,
                        'theme': theme_name
                    })
                    self.all_frame_data.append(frame)
                sample_idx += 1
        
        # 按Sample划分测试集（和训练时划分逻辑一致，确保无数据泄露）
        all_sample_ids = list(set([info['sample_idx'] for info in self.all_frame_info]))
        _, test_sample_ids = train_test_split(
            all_sample_ids, test_size=test_ratio, random_state=42  # 固定随机种子，和训练时一致
        )
        
        # 筛选测试集数据
        self.frame_data = []
        self.frame_info = []
        for idx, info in enumerate(self.all_frame_info):
            if info['sample_idx'] in test_sample_ids:
                self.frame_data.append(self.all_frame_data[idx])
                self.frame_info.append(info)
        
        self.frame_data = np.array(self.frame_data, dtype=np.float32)
        self.frame_info_df = pd.DataFrame(self.frame_info)
        self.sample_count = len(test_sample_ids)
        
        print(f"测试集 - Theme：{target_theme} | 帧数：{len(self.frame_data)} | Sample数：{self.sample_count}")

    def __len__(self):
        return len(self.frame_data)
    
    def __getitem__(self, idx):
        return torch.tensor(self.frame_data[idx], dtype=torch.float32)

# ========== 加载模型结构（和训练时完全一致） ==========
class FrameAutoencoder(nn.Module):
    """帧级别骨架自编码器（和训练代码完全相同）"""
    def __init__(self, input_channels=3, num_joints=71, hidden_dim=128):
        super().__init__()
        self.input_dim = num_joints * input_channels
        self.hidden_dim = hidden_dim
        
        self.encoder = nn.Sequential(
            nn.Flatten(),
            nn.Linear(self.input_dim, hidden_dim * 2),
            nn.LayerNorm(hidden_dim * 2),
            nn.GELU(),
            nn.Dropout(0.1),
            nn.Linear(hidden_dim * 2, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.GELU(),
            nn.Linear(hidden_dim, hidden_dim // 2)
        )
        
        self.decoder = nn.Sequential(
            nn.Linear(hidden_dim // 2, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.GELU(),
            nn.Dropout(0.1),
            nn.Linear(hidden_dim, hidden_dim * 2),
            nn.LayerNorm(hidden_dim * 2),
            nn.GELU(),
            nn.Linear(hidden_dim * 2, self.input_dim),
            nn.Unflatten(dim=1, unflattened_size=(num_joints, input_channels))
        )
    
    def forward(self, x):
        z = self.encoder(x)
        x_recon = self.decoder(z)
        return x_recon
    
    def get_reconstruction_error(self, x):
        """计算重建误差（MSE）"""
        with torch.no_grad():
            x_recon = self.forward(x)
            mse = torch.mean((x - x_recon) ** 2, dim=[1, 2])
        return mse.cpu().numpy()

# ========== 核心测试函数（加载模型+推理+计算结果） ==========
def load_trained_model(theme):
    """加载指定Theme的训练好的模型权重"""
    model_path = os.path.join(OUTPUT_DIR, f"autoencoder_{theme}_{MODALITY}.pth")
    if not os.path.exists(model_path):
        raise FileNotFoundError(f"未找到{theme}的模型权重：{model_path}")
    
    # 初始化模型
    input_channels = 2 if MODALITY == '2d' else 3
    model = FrameAutoencoder(
        input_channels=input_channels,
        num_joints=NUM_JOINTS,
        hidden_dim=128
    ).to(device)
    
    # 加载权重
    checkpoint = torch.load(model_path)
    model.load_state_dict(checkpoint['model_state_dict'])
    model.eval()  # 切换到评估模式
    print(f"成功加载模型：{model_path}")
    return model, checkpoint['mean'], checkpoint['std']

def test_single_theme(theme, pkl_files):
    """测试单个Theme"""
    # 1. 创建该Theme的测试集
    test_dataset = AnomalySkeletonTestDataset(
        pkl_files=pkl_files,
        modality=MODALITY,
        target_theme=theme,
        max_frames=MAX_FRAMES,
        test_ratio=TEST_SPLIT_RATIO
    )
    if len(test_dataset) == 0:
        print(f"警告：{theme}测试集为空，跳过")
        return None, None, None
    
    # 2. 加载训练好的模型
    model, train_mean, train_std = load_trained_model(theme)
    
    # 3. 测试集数据归一化（使用训练集的均值和标准差，避免数据泄露）
    test_dataset.frame_data = (test_dataset.frame_data - train_mean) / train_std
    
    # 4. 模型推理，计算重建误差
    test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)
    all_errors = []
    for batch in tqdm(test_loader, desc=f"推理 {theme} 测试集"):
        batch = batch.to(device)
        errors = model.get_reconstruction_error(batch)
        all_errors.extend(errors)
    
    # 5. 计算异常概率（优化逻辑，让分布集中在0/1）
    frame_info_df = test_dataset.frame_info_df.copy()
    frame_info_df['reconstruction_error'] = all_errors
    
    # 平方+对数放大差异
    log_errors = np.log1p(np.square(frame_info_df['reconstruction_error'].values))
    scaler = MinMaxScaler(feature_range=(0, 10))
    frame_info_df['anomaly_score'] = scaler.fit_transform(log_errors.reshape(-1, 1))
    
    # Sigmoid让概率向0/1集中
    frame_info_df['anomaly_prob'] = 1 / (1 + np.exp(-frame_info_df['anomaly_score']))
    
    # 6. 判定异常帧
    anomaly_threshold = np.quantile(frame_info_df['anomaly_prob'], ANOMALY_THRESHOLD)
    frame_info_df['is_anomaly'] = frame_info_df['anomaly_prob'] > anomaly_threshold
    
    # 7. 统计测试结果
    test_sample_stats = frame_info_df.groupby('sample_idx').agg({
        'motion_name': 'first',
        'frame_idx': ['min', 'max', 'count'],
        'is_anomaly': ['sum', 'mean'],
        'anomaly_prob': ['mean', 'max']
    }).round(4)
    test_sample_stats.columns = ['motion_name', '最小帧索引', '最大帧索引', '总帧数', 
                                '异常帧数', '异常帧比例', '平均异常概率', '最大异常概率']
    test_sample_stats['异常帧比例'] = test_sample_stats['异常帧比例'] * 100
    
    return frame_info_df, anomaly_threshold, test_sample_stats

# ========== 执行纯测试（主流程） ==========
if __name__ == "__main__":
    # 1. 加载PKL文件和所有Theme
    pkl_files = get_pkl_files(PKL_2D_DIR) if MODALITY == '2d' else get_pkl_files(PKL_3D_DIR)
    all_themes = get_all_themes(pkl_files)
    print(f"\n找到{len(all_themes)}个Theme：{all_themes}")
    
    # 2. 逐个Theme测试
    theme_test_results = {}
    for theme in all_themes:
        print(f"\n" + "="*80)
        print(f"开始测试 Theme：{theme}")
        print("="*80)
        try:
            test_frame_df, threshold, test_stats = test_single_theme(theme, pkl_files)
            if test_frame_df is not None:
                theme_test_results[theme] = (test_frame_df, threshold, test_stats)
                
                # 保存该Theme的测试结果
                test_frame_df.to_csv(os.path.join(OUTPUT_DIR, f"{theme}_test_frame_level_results.csv"), index=False, encoding='utf-8-sig')
                test_stats.to_csv(os.path.join(OUTPUT_DIR, f"{theme}_test_sample_anomaly_stats.csv"), encoding='utf-8-sig')
                print(f"{theme} 测试结果已保存")
        except Exception as e:
            print(f"测试{theme}失败：{str(e)}")
            continue
    
    # 3. 测试结果可视化（每行2个子图）
    print(f"\n" + "="*80)
    print(f"测试结果可视化")
    print("="*80)
    n_cols = 2
    n_themes = len(theme_test_results)
    n_rows = (n_themes + n_cols - 1) // n_cols
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 6*n_rows))
    plt.suptitle(f'{MODALITY}模态 - 测试集异常概率分布（纵坐标=帧数）', fontsize=16, y=0.98)
    
    if n_rows == 1:
        axes = axes.reshape(1, -1)
    
    # 绘制每个Theme的测试集分布
    for idx, (theme, (test_frame_df, threshold, _)) in enumerate(theme_test_results.items()):
        row = idx // n_cols
        col = idx % n_cols
        ax = axes[row, col]
        
        anomaly_prob = test_frame_df['anomaly_prob'].values
        total_frames = len(test_frame_df)
        anomaly_frames = len(test_frame_df[test_frame_df['is_anomaly']])
        anomaly_ratio = (anomaly_frames / total_frames) * 100
        
        # 直方图
        n, bins, patches = ax.hist(anomaly_prob, bins=30, color='#2ca02c', alpha=0.8, edgecolor='black')
        # 高亮异常帧区间
        for i, patch in enumerate(patches):
            if bins[i] >= threshold:
                patch.set_facecolor('#ff4444')
        
        # 图表配置
        ax.set_xlabel('异常概率')
        ax.set_ylabel('帧数')
        ax.set_xlim(0, 1.0)
        ax.set_ylim(0, max(n) * 1.1)
        ax.grid(axis='y', alpha=0.3)
        ax.axvline(x=threshold, color='red', linestyle='--', label=f'阈值={threshold:.3f}')
        ax.legend()
        ax.set_title(f'Theme: {theme}\n帧数：{total_frames} | 异常帧：{anomaly_frames}（{anomaly_ratio:.1f}%）')
    
    # 隐藏多余子图
    for idx in range(n_themes, n_rows*n_cols):
        axes[idx//n_cols, idx%n_cols].set_visible(False)
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.93)
    plt.savefig(os.path.join(OUTPUT_DIR, 'test_set_anomaly_distribution.png'), dpi=300, bbox_inches='tight')
    plt.show()
    
    # 4. 输出测试汇总
    print(f"\n测试完成！所有结果已保存到 {OUTPUT_DIR}")
    print(f"测试成功的Theme：{list(theme_test_results.keys())}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import pickle
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, roc_curve
import scipy.stats as stats
import seaborn as sns
from tqdm import tqdm

# ========== 全局配置（与训练代码一致） ==========
OUTPUT_DIR = 'asd_anomaly_detection_results_frame_level'
MODALITY = '3d'  # 与训练时一致
NUM_JOINTS = 71  # 与训练时一致
PKL_DIR = '/root/second/3D'  # 与训练时的PKL路径一致（用于骨架可视化）

# 字体配置（使用系统自带字体，避免中文乱码）
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei', 'Arial Unicode MS']  # 兼容不同系统
plt.rcParams['axes.unicode_minus'] = False

# 骨骼连接（71关节默认配置，可根据实际关节调整）
SKELETON_CONNECTIONS = [
    (0, 1), (1, 2), (2, 3), (3, 4),  # 头部
    (0, 5), (5, 6), (6, 7), (7, 8),  # 左肩-左臂
    (0, 9), (9, 10), (10, 11), (11, 12),  # 右肩-右臂
    (0, 13), (13, 14), (14, 15), (15, 16),  # 躯干-左腿
    (0, 17), (17, 18), (18, 19), (19, 20)   # 躯干-右腿
]

# ========== 自动加载所有Theme结果 ==========
def get_all_themes():
    """从结果目录提取所有Theme名称"""
    themes = []
    for file in os.listdir(OUTPUT_DIR):
        if file.endswith('_frame_level_results.csv'):
            theme = file.replace('_frame_level_results.csv', '')
            themes.append(theme)
    return sorted(list(set(themes)))

all_themes = get_all_themes()
assert len(all_themes) > 0, f"未在{OUTPUT_DIR}找到结果文件"

# 加载每个Theme的逐帧结果和统计信息
theme_data = {}  # {theme: frame_df}
for theme in all_themes:
    frame_csv = os.path.join(OUTPUT_DIR, f"{theme}_frame_level_results.csv")
    frame_df = pd.read_csv(frame_csv, encoding='utf-8-sig')
    theme_data[theme] = frame_df

print(f"自动识别到{len(all_themes)}个Theme：{all_themes}")
print("所有结果文件加载完成！")
```

```python
# 计算子图行数（每行2个，向上取整）
n_themes = len(all_themes)
n_cols = 2
n_rows = (n_themes + n_cols - 1) // n_cols

# 创建画布（调整尺寸适配每行2个子图）
fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 6*n_rows))
plt.suptitle('All Themes - Anomaly Frame Continuity Verification', fontsize=16, y=0.98)

# 处理单一行的情况（确保axes是二维数组）
if n_rows == 1:
    axes = axes.reshape(1, -1)

for idx, theme in enumerate(all_themes):
    row = idx // n_cols
    col = idx % n_cols
    ax = axes[row, col]
    
    frame_df = theme_data[theme]
    # 选择前5个有异常帧的Sample（避免拥挤）
    anomaly_samples = frame_df[frame_df['is_anomaly']]['sample_idx'].unique()[:5]
    if len(anomaly_samples) < 1:
        anomaly_samples = frame_df['sample_idx'].unique()[:3]
    
    # 绘制每个Sample的异常帧分布
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
    for i, sample_id in enumerate(anomaly_samples):
        sample_frames = frame_df[frame_df['sample_idx'] == sample_id].sort_values('frame_idx')
        frame_indices = sample_frames['frame_idx'].values
        is_anomaly = sample_frames['is_anomaly'].values
        anomaly_prob = sample_frames['anomaly_prob'].values
        
        # 正常帧（浅色点）
        normal_mask = ~is_anomaly
        ax.scatter(frame_indices[normal_mask], anomaly_prob[normal_mask], 
                   c=colors[i], s=15, alpha=0.4, label=f'Sample {sample_id}' if i==0 else "")
        
        # 异常帧（深色三角形）
        anomaly_mask = is_anomaly
        ax.scatter(frame_indices[anomaly_mask], anomaly_prob[anomaly_mask], 
                   c=colors[i], s=40, alpha=0.8, marker='^')
        
        # 标注连续异常段（阴影）
        if anomaly_mask.any():
            anomaly_runs = []
            start = None
            for f_idx, (f, is_a) in enumerate(zip(frame_indices, anomaly_mask)):
                if is_a and start is None:
                    start = f
                elif not is_a and start is not None:
                    anomaly_runs.append((start, frame_indices[f_idx-1]))
                    start = None
            if start is not None:
                anomaly_runs.append((start, frame_indices[-1]))
            # 绘制阴影
            for (s, e) in anomaly_runs:
                ax.axvspan(s, e, alpha=0.15, color=colors[i])
    
    # 图配置
    ax.set_xlabel('Frame Index')
    ax.set_ylabel('Anomaly Probability')
    ax.set_title(f'Theme: {theme}\nTop {len(anomaly_samples)} Samples', fontsize=11)
    ax.set_ylim(-0.05, 1.05)
    ax.grid(alpha=0.3)
    ax.legend(fontsize=8)

# 隐藏多余子图
for idx in range(n_themes, n_rows*n_cols):
    row = idx // n_cols
    col = idx % n_cols
    axes[row, col].set_visible(False)

plt.tight_layout()
plt.subplots_adjust(top=0.93)
plt.savefig(os.path.join(OUTPUT_DIR, 'all_themes_anomaly_continuity.png'), dpi=300, bbox_inches='tight')
plt.show()

# 输出连续性统计
print("="*50)
print("Anomaly Frame Continuity Statistics (Threshold: Consecutive ≥ 2 frames)")
print("="*50)
for theme in all_themes:
    frame_df = theme_data[theme]
    anomaly_samples = frame_df[frame_df['is_anomaly']]['sample_idx'].unique()
    total_anomaly_frames = len(frame_df[frame_df['is_anomaly']])
    consecutive_anomaly_frames = 0
    
    for sample_id in anomaly_samples:
        sample_frames = frame_df[frame_df['sample_idx'] == sample_id].sort_values('frame_idx')
        anomaly_mask = sample_frames['is_anomaly'].values
        # 计算连续异常帧数量
        runs = []
        start = None
        for i, is_a in enumerate(anomaly_mask):
            if is_a and start is None:
                start = i
            elif not is_a and start is not None:
                runs.append(i - start)
                start = None
        if start is not None:
            runs.append(len(anomaly_mask) - start)
        # 累计连续段≥2帧的异常帧
        consecutive_anomaly_frames += sum(r for r in runs if r >= 2)
    
    if total_anomaly_frames > 0:
        consecutive_ratio = (consecutive_anomaly_frames / total_anomaly_frames) * 100
    else:
        consecutive_ratio = 0.0
    
    print(f"\n{theme}:")
    print(f"  Total Anomaly Frames: {total_anomaly_frames}")
    print(f"  Consecutive Anomaly Frames (≥2): {consecutive_anomaly_frames}")
    print(f"  Consecutive Ratio: {consecutive_ratio:.1f}% (≥80% is reasonable)")
```

```python
# 辅助函数：从PKL加载骨架数据
def load_skeleton(theme, motion_name):
    """根据Theme和动作名称加载原始骨架"""
    pkl_files = [f for f in os.listdir(PKL_DIR) if f.endswith('.pkl')]
    for pkl_file in pkl_files:
        with open(os.path.join(PKL_DIR, pkl_file), 'rb') as f:
            data = pickle.load(f)
        if data['metadata']['theme_name'] != theme:
            continue
        for sample in data['samples']:
            if sample['motion_name'] == motion_name:
                skeleton_key = 'skeleton_2d' if MODALITY == '2d' else 'skeleton_3d'
                return sample[skeleton_key]  # (T, J, C)
    return None

# 辅助函数：绘制单帧骨架
def plot_skeleton(ax, x, y, color, label):
    ax.scatter(x, y, c=color, s=40, alpha=0.8, label=label)
    for (parent, child) in SKELETON_CONNECTIONS:
        if parent < len(x) and child < len(x):
            ax.plot([x[parent], x[child]], [y[parent], y[child]], c=color, linewidth=2, alpha=0.6)
    ax.invert_yaxis()
    ax.set_aspect('equal')

# 计算子图行数（每行2个）
n_rows = (n_themes + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 7*n_rows))
plt.suptitle('All Themes - Skeleton Comparison (Anomaly vs Normal Frames)', fontsize=16, y=0.98)

if n_rows == 1:
    axes = axes.reshape(1, -1)

for idx, theme in enumerate(all_themes):
    row = idx // n_cols
    col = idx % n_cols
    ax = axes[row, col]
    
    frame_df = theme_data[theme]
    # 找到第一个有异常帧的Sample
    anomaly_samples = frame_df[frame_df['is_anomaly']]['sample_idx'].unique()
    if len(anomaly_samples) == 0:
        ax.text(0.5, 0.5, f'Theme: {theme}\nNo Anomaly Frames', 
                ha='center', va='center', transform=ax.transAxes, fontsize=12)
        ax.set_xlim(0, 1)
        ax.set_ylim(0, 1)
        continue
    
    sample_id = anomaly_samples[0]
    sample_frames = frame_df[frame_df['sample_idx'] == sample_id].sort_values('frame_idx')
    anomaly_frame_indices = sample_frames[sample_frames['is_anomaly']]['frame_idx'].values
    # 取异常段中间帧+前后正常帧
    anomaly_frame_idx = anomaly_frame_indices[len(anomaly_frame_indices)//2]
    normal_frame1_idx = max(0, anomaly_frame_idx - 1)
    normal_frame2_idx = min(len(sample_frames)-1, anomaly_frame_idx + 1)
    
    # 加载骨架数据
    motion_name = sample_frames['motion_name'].iloc[0]
    skeleton_data = load_skeleton(theme, motion_name)
    if skeleton_data is None:
        ax.text(0.5, 0.5, f'Theme: {theme}\nSkeleton Data Not Found', 
                ha='center', va='center', transform=ax.transAxes, fontsize=12)
        continue
    
    # 提取坐标（3D取x,y投影）
    def get_coords(f_idx):
        frame = skeleton_data[f_idx]
        return frame[:, 0], frame[:, 1]
    
    x_norm1, y_norm1 = get_coords(normal_frame1_idx)
    x_anom, y_anom = get_coords(anomaly_frame_idx)
    x_norm2, y_norm2 = get_coords(normal_frame2_idx)
    
    # 绘制对比骨架
    plot_skeleton(ax, x_norm1, y_norm1, 'green', f'Normal {normal_frame1_idx}')
    plot_skeleton(ax, x_anom, y_anom, 'red', f'Anomaly {anomaly_frame_idx}')
    plot_skeleton(ax, x_norm2, y_norm2, 'blue', f'Normal {normal_frame2_idx}')
    
    # 图配置
    ax.set_title(f'Theme: {theme}\nSample {sample_id}', fontsize=12)
    ax.legend(loc='upper right', fontsize=9)
    ax.grid(alpha=0.3)

# 隐藏多余子图
for idx in range(n_themes, n_rows*n_cols):
    row = idx // n_cols
    col = idx % n_cols
    axes[row, col].set_visible(False)

plt.tight_layout()
plt.subplots_adjust(top=0.93)
plt.savefig(os.path.join(OUTPUT_DIR, 'all_themes_skeleton_comparison.png'), dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.preprocessing import MinMaxScaler

# ========== 核心配置 ==========
OUTPUT_DIR = 'asd_anomaly_detection_results_frame_level'
ANOMALY_THRESHOLD = 0.8  # 每个Theme独立计算阈值
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False

# 自动获取所有Theme
def get_all_themes():
    themes = []
    for file in os.listdir(OUTPUT_DIR):
        if file.endswith('_frame_level_results.csv'):
            themes.append(file.replace('_frame_level_results.csv', ''))
    return sorted(list(set(themes)))

all_themes = get_all_themes()

# ========== 按Theme独立计算异常概率（每帧的异常可能性） ==========
def compute_theme_anomaly_prob(frame_df):
    """每个Theme独立处理：计算该Theme内每帧的异常概率"""
    # 1. 对该Theme的重建误差取对数（放大差异）
    log_errors = np.log1p(frame_df['reconstruction_error'].values)  # 避免log(0)
    # 2. 归一化到0-5（该Theme内独立归一化）
    scaler = MinMaxScaler(feature_range=(0, 5))
    scaled_errors = scaler.fit_transform(log_errors.reshape(-1, 1)).flatten()
    # 3. Sigmoid函数：该Theme内每帧的异常概率（0→正常，1→异常）
    anomaly_prob = 1 / (1 + np.exp(-scaled_errors))
    # 4. 该Theme独立的异常阈值
    threshold = np.quantile(anomaly_prob, ANOMALY_THRESHOLD)
    is_anomaly = anomaly_prob > threshold
    return anomaly_prob, threshold, is_anomaly

# ========== 按Theme绘制独立的直方图（每个Theme一个图） ==========
# 计算子图布局（每行2个Theme，适配数量）
n_cols = 2
n_rows = (len(all_themes) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 6*n_rows))
plt.suptitle('Per-Theme Anomaly Probability Distribution\n(Each Frame\'s Anomaly Likelihood)', fontsize=16, y=0.98)

# 处理单一行的情况
if n_rows == 1:
    axes = axes.reshape(1, -1)

for idx, theme in enumerate(all_themes):
    row = idx // n_cols
    col = idx % n_cols
    ax = axes[row, col]
    
    # 加载该Theme的原始数据
    frame_csv = os.path.join(OUTPUT_DIR, f"{theme}_frame_level_results.csv")
    frame_df = pd.read_csv(frame_csv, encoding='utf-8-sig')
    
    # 计算该Theme内每帧的异常概率
    anomaly_prob, threshold, is_anomaly = compute_theme_anomaly_prob(frame_df)
    
    # 绘制该Theme的直方图（纵坐标=该Theme的帧数）
    n, bins, patches = ax.hist(
        anomaly_prob,
        bins=30,
        color='#2ca02c',
        alpha=0.8,
        edgecolor='black',
        linewidth=0.5
    )
    
    # 高亮该Theme的异常帧（>阈值的区间标红）
    for i, patch in enumerate(patches):
        if bins[i] >= threshold:
            patch.set_facecolor('#ff4444')
    
    # 图配置（每个Theme独立）
    ax.set_xlabel('Anomaly Probability (Per Frame)', fontsize=10)
    ax.set_ylabel('Frame Count (This Theme)', fontsize=10)  # 纵坐标是该Theme的帧数
    ax.set_xlim(0, 1.0)
    ax.set_ylim(0, max(n) * 1.1)
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    ax.axvline(x=threshold, color='red', linestyle='--', linewidth=2, 
               label=f'Threshold = {threshold:.3f}')
    ax.legend(fontsize=9)
    ax.set_title(f'Theme: {theme}\nAnomaly Frames: {sum(is_anomaly)} ({sum(is_anomaly)/len(frame_df)*100:.1f}%)', 
                 fontsize=11)

# 隐藏多余子图
for idx in range(len(all_themes), n_rows*n_cols):
    row = idx // n_cols
    col = idx % n_cols
    axes[row, col].set_visible(False)

plt.tight_layout()
plt.subplots_adjust(top=0.93)
plt.savefig(os.path.join(OUTPUT_DIR, 'per_theme_anomaly_prob_distribution.png'), dpi=300, bbox_inches='tight')
plt.show()

# ========== 保存每个Theme的逐帧异常概率（可选） ==========
for theme in all_themes:
    frame_csv = os.path.join(OUTPUT_DIR, f"{theme}_frame_level_results.csv")
    frame_df = pd.read_csv(frame_csv, encoding='utf-8-sig')
    anomaly_prob, threshold, is_anomaly = compute_theme_anomaly_prob(frame_df)
    # 新增列：该Theme内每帧的异常概率、是否异常
    frame_df['anomaly_prob_per_theme'] = anomaly_prob
    frame_df['is_anomaly_per_theme'] = is_anomaly
    frame_df.to_csv(os.path.join(OUTPUT_DIR, f"{theme}_frame_level_results_per_theme.csv"), 
                   index=False, encoding='utf-8-sig')

print("所有Theme的逐帧异常概率已独立计算并保存！")
```
